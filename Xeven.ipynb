```python
import asyncio
import re
from crawl4ai import AsyncWebCrawler, CrawlerRunConfig

async def main():
    config = CrawlerRunConfig(
        excluded_tags=["form", "header", "footer"],
        keep_data_attributes=False
    )
    async with AsyncWebCrawler() as crawler:
        result = await crawler.arun("https://techcrunch.com/latest/", config=config)

        pattern = r"https://techcrunch.com/\d{4}/\d{2}/\d{2}/.*"
        filtered_links = [link['href'] for link in result.links['internal'] if isinstance(link, dict) and 'href' in link and re.match(pattern, link['href'])]

        # Process only the first 10 links
        for i, link in enumerate(filtered_links[:10]):
            print(f"Link number {i + 1} result")  # Corrected the printing
            link_result = await crawler.arun(link, config=config)
            if i == 1:
                print(result.markdown)  # Print result.markdown when i is 1
            else:
                print(link_result.markdown[:300])

if __name__ == "__main__":
    asyncio.run(main())
```
